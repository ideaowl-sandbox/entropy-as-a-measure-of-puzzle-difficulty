```python
import requests
import json
import time
import random
```

```python
start_url = 'https://windmill.thefifthmatt.com/_/things?sort=new'
```

```python
puzzles = []

def get_puzzles_by_url(start_id=None):
    curr_url = start_url
    if start_id is not None:
        curr_url += '&start=' + start_id
    server_reply = requests.get(curr_url)
    puzzles_meta = json.loads(server_reply.content)
    response = {'puzzles': puzzles_meta['things'], 'has_more': puzzles_meta['hasMore']}
    return response

def obtain_puzzles():
    has_more = True
    start_id = None
    num_requests = 0
    while (has_more is True): # and num_requests < 3 
        time.sleep(random.randrange(4, 11, 1) / 10)
        response = get_puzzles_by_url(start_id)
        has_more = response['has_more']
        start_id = response['puzzles'][len(response['puzzles']) - 1]['id']
        puzzles.extend(response['puzzles'])
        print('Number of Puzzles: ', len(puzzles))
        num_requests += 1
        if (num_requests % 10 == 0):
            save_file()

def save_file():
    with open('windmill-puzzles.json', 'w') as outfile:
        json.dump(puzzles, outfile)
            
obtain_puzzles()
save_file()
```
